```python
#Import LIBRARIES

import matplotlib.pyplot as plt
import pandas as pd
from datasets import load_dataset

#Loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#data clean up

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
date_counts =df.job_posted_date.value_counts() #IT PROVIDES THE NUMBER OF DIFFRENT JOB POSTING IN THE ASSOCIETED DATETIME

date_counts =date_counts.sort_index()
```

```python
#              x                 y
plt.plot(date_counts.index, date_counts)
plt.show()
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.month

monthly_counts = df. job_posted_month.value_counts()

monthly_counts = monthly_counts.sort_index()
```

```python
plt.plot(monthly_counts.index, monthly_counts.values)
plt.show()
```

```python
df['job_posted_day'] = df['job_posted_date'].dt.day
```

```python
dayly_counts = df.job_posted_day.value_counts()

dayly_counts = dayly_counts.sort_index()

dayly_counts
```

```python
#Create a horizontal bar chart showing the average yearly salary for different job titles. Use the job_title_short and salary_year_avg 
# columns to calculate the average salary for each job title.
```

```python
job_salary = df.groupby('job_title_short')['salary_year_avg'].mean()
job_salary= job_salary.sort_values(ascending=True)
```

```python
plt.barh(job_salary.index, job_salary.values)
```
