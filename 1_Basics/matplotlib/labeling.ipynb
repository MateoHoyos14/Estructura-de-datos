```python
#Import LIBRARIES

import matplotlib.pyplot as plt
import pandas as pd
from datasets import load_dataset

#Loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#data clean up

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
job_counts = df.job_title_short.value_counts()
```

```python
plt.bar(job_counts.index, job_counts, color ='purple')
plt.title('Posting by Job Title')
plt.ylabel('Count of Job Postings')
plt.xticks(rotation = 45, ha='right')
plt.show()
```

```python
number_postings = df.job_country.value_counts().head(5)
```

```python
plt.bar(number_postings.index, number_postings, color = 'orange')

plt.title('Number of Jobs per Country')
plt.ylabel('Number of Postings')
plt.xlabel('Country')
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

```python
avg_salary = df.groupby('job_title_short')['salary_year_avg'].mean()
avg_salary = avg_salary.sort_values(ascending=True)
```

```python
plt.barh(avg_salary.index, avg_salary, color = 'green')
plt.title('Average Salary by Job Title')
plt.xlabel('Average Salary')
plt.ylabel('JOB TITLE')

plt.show()
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.month

salary_month = df.groupby('job_posted_month')['salary_year_avg'].mean().sort_index()

plt.plot(salary_month.index, salary_month, color='red', label='Average Salary')
plt.title('Average Salary by Job Month')
plt.xlabel('Month')
plt.ylabel('Average Salary')
plt.legend()
plt.show()
```
