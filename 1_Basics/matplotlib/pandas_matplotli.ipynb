```python
#Import LIBRARIES

import matplotlib.pyplot as plt
import pandas as pd
from datasets import load_dataset

#Loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#data clean up

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
job_counts = df.job_title_short.value_counts()

plt.bar(job_counts.index, job_counts)
```

```python
job_counts.plot(kind ='bar', color = 'purple')
plt.title('Posting by Job Title')
plt.xlabel('')
plt.ylabel('Count of Job Postings')
plt.xticks(rotation = 45, ha='right')
plt.show()
```

```python
df.plot(x='job_posted_date', y ='salary_year_avg', kind = 'line')
```

```python
job_platforms = df['job_via'].value_counts().head(5)


job_platforms.plot(kind = 'bar')
```

```python
df['monthly_job_counts'] = df['job_posted_date'].dt.to_period('M')

monthly =df.monthly_job_counts.value_counts().sort_index()

monthly.plot(x= ' ', y=' ')
plt.title('Trend of Job Postings Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Job Postings')
plt.show()
```

```python
location = df['job_location'].value_counts().head(5)

location.plot(kind = 'barh', color='blue')

plt.title('Number of Job Postings by Job Location')
plt.xlabel('Count of Jobs')
plt.ylabel('Job Location')

plt.show()
```
